```python
data3 = pd.read_excel(r'./rpt_share.xlsx',index_col=0)
data3
```

```python
year = data3.index
a = data3['a']
h = data3['h']
x = data3['x']

lenth = pd.np.arange(len(year))
width = 0.25
fig, ax = plt.subplots()
fig.set_size_inches(16, 8)
v1 = ax.bar(lenth - width, x , width, color='#75bbfd',label='非个股报告')
v2 = ax.bar(lenth - 0.1, a , width, color='#3c73a8',label='A股报告',tick_label = year)
v3 = ax.bar(lenth + width - 0.2, h , width, color='#0a437a',label='H股报告')
plt.xticks(fontsize =16)
plt.yticks(fontsize =16)
plt.legend(loc='best',fontsize='x-large')
for rect in v1 +v2+ v3:
    h= rect.get_height()
    ax.text(rect.get_x()+rect.get_width()/2 , h, '%d' % int(h),ha = 'center', va = 'bottom',size=14)
    
plt.show()
```

```python
data = {
    '< 30分钟':(1901, '#95a5a6'),
    '>= 30分钟':(84, '#e74c3c'),
}
# 设置绘图对象的大小
fig, ax = plt.subplots()
fig.set_size_inches(16, 8)

frame = plt.gca()
# y 轴不可见
frame.axes.get_yaxis().set_visible(False)
# x 轴不可见
frame.axes.get_xaxis().set_visible(False)

types = data.keys()
values = [x[0] for x in data.values()]
colors = [x[1] for x in data.values()]
    
ax1 = fig.add_subplot(121)
ax1.set_title('1月超过30分钟占比',size=18)

labels = ['{}:{}'.format(t, v) for t, v in zip(types,values)]

# 设置饼图的凸出显示
explode = [0, 0.2 ]

# 画饼状图， 并且指定标签和对应的颜色
# 指定阴影效果
ax1.pie(values, labels = labels, colors=colors, explode=explode, shadow=True,startangle=-30,autopct='%1.1f%%'
                               ,textprops={'fontsize':16})

#labeldistance，文本的位置离远点有多远，1.1指1.1倍半径的位置
#autopct，圆里面的文本格式，%3.1f%%表示小数有三位，整数有一位的浮点数
#shadow，饼是否有阴影
#startangle，起始角度，0，表示从0开始逆时针转，为第一块。一般选择从90度开始比较好看
#pctdistance，百分比的text离圆心的距离
#patches, l_texts, p_texts，为了得到饼图的返回值，p_texts饼图内部文本的，l_texts饼图外label的文本

#改变文本的大小
#方法是把每一个text遍历。调用set_size方法设置它的属性

# 设置x，y轴刻度一致，这样饼图才能是圆的
ax1.axis('equal')
plt.legend(fontsize=12,loc='upper left')


data = {
    '工作时间':(54, '#e74c3c'),
    '非工作时间':(30, '#95a5a6'),
}

types2 = data.keys()
values2 = [x[0] for x in data.values()]
colors2 = [x[1] for x in data.values()]

ax2 = fig.add_subplot(122)
ax2.set_title('工作时间超过30分钟占比',size=18)
# ax2.set_size(6,6)

labels2 = ['{}:{}'.format(t, v) for t, v in zip(types2,values2)]

# 设置饼图的凸出显示
explode2 = [0, 0.1 ]

# 画饼状图， 并且指定标签和对应的颜色
# 指定阴影效果
ax2.pie(values2, labels = labels2, colors=colors2, explode=explode2, shadow=True,startangle=90,
                                autopct='%1.1f%%',textprops={'fontsize':16})

plt.legend(fontsize=12,loc='upper right')

ax2.axis('equal')
plt.show()
```
